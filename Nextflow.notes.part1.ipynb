```python
print('''

Inputs:

The input block allows you to define the INPUT CHANNELS of a process, similar to FUNCTION ARGUMENTS. 

input:
  <input qualifier> <input name>
  
An input definition consists of a QUALIFIER and a NAME. The input qualifier defines the type of data to be received.

The following input qualifiers are available:

val: Access the input value by name in the process script.

path: Handle the input value as a path, staging the file properly in the execution context.

env: Use the input value to set an environment variable in the process script.

stdin: Forward the input value to the process stdin special file.

tuple: Handle a group of input values having any of the above qualifiers.

each: Execute the process for each element in the input collection.

''')
```

```python
print("""
CLOSURES in Python : 

In Python, a closure is a function that retains access to the variables in its enclosing scope, even after that scope has exited. 
Closures are a key feature of Python's support for LEXICAL SCOPING.

What is a Closure?

Enclosing Function: A closure is created when a nested function (inner function) captures variables from its enclosing function.
Captured Variables: These variables are "remembered" by the closure, even after the enclosing function has finished execution.
Use Case: Closures are useful for scenarios where you want to encapsulate and maintain state across function calls without using 
global variables or classes.
""")

def outer_function(x):
    def inner_function(y):
        return x + y  # x is captured from the outer function's scope
    return inner_function

closure = outer_function(10)  # x = 10
print(closure(5))  # Output: 15
print(closure(20))  # Output: 30

# outer_function returns inner_function.
# The x variable is captured by inner_function, creating a closure.

def multiplier(factor):
    def multiply_by(n):
        return n * factor
    return multiply_by

double = multiplier(2)
triple = multiplier(3)

print(double(5))  # Output: 10
print(triple(5))  # Output: 15

print('''

Practical Use Cases for Closures :

Data Encapsulation: Closures can encapsulate and protect data from external modification.
Callback Functions: Closures are often used in callback-based programming (e.g., event handling).
Decorators: Python decorators heavily rely on closures to modify or extend the behavior of functions.

''')

def decorator(func):
    def wrapper(*args, **kwargs):
        print("Before the function call")
        result = func(*args, **kwargs)
        print("After the function call")
        return result
    return wrapper

@decorator
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")
# Output:
# Before the function call
# Hello, Alice!
# After the function call

print('''Closures vs Objects :

Closures can often achieve similar functionality as objects (maintaining state and behavior). 
However, closures are more lightweight and are typically preferred when only a small amount 
of state needs to be maintained.
''')
```
